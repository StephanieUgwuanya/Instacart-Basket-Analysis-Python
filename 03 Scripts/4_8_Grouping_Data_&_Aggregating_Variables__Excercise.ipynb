# 4. Creating a loyalty flag for existing customers using the transform() and loc() functions

```python
# Step 1

# Aggregating Data with transform()

# new column created with the max orders
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
# Step 2

# Deriving Columns with loc()

#Criteria

# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Step 3

# Check to see the flags have been applied correctly
```

```python
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# 6. Create a spending flag for each user based on the average price across all their orders.

```python
# Step 1

# Aggregating Data with transform()

# new column created with the mean price
df_ords_prods_merge['mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
df_ords_prods_merge.head(5) #checking if the new column mean_price has been generated
```

```python
# Step 2

# Deriving Columns with loc()

#Criteria

# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] < 10, 'spending flag'] = 'Low Spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] >= 10, 'spending flag'] = 'Higher Spender'
```

```python
df_ords_prods_merge.rename(columns={'spending flag': 'spending_flag'}, inplace=True) #changed the column name to include the underscore
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Step 3

# Check to see the flags have been applied correctly

df_ords_prods_merge[['user_id', 'spending_flag', 'mean_price']].head(60)

```

# 7. Create an order frequency flag

the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Step 1

# Aggregating Data with transform()

# new column created with the mean price
df_ords_prods_merge['customer_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
df_ords_prods_merge.head(5)
```

```python
# Step 2

# Deriving Columns with loc()

# Criteria

# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['customer_frequency'] >10) & (df_ords_prods_merge['customer_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Step 3

# Check to see the flags have been applied correctly

df_ords_prods_merge[['user_id', 'frequency_flag', 'customer_frequency']].head(60)
```
