# 4. Merging df_ords_prods_combined and df_prods

```python
# As the data shapes are completely different for df_ords_prods_combined and df_prods I will use a merge to combine the dataframes on product_id
```

```python
#we have a column called merge in df_ords_prods_combined which is likely effecting the syntax, this will be removed

print(df_ords_prods_combined.columns)
```

```python
df_ords_prods_combined = df_ords_prods_combined.drop(columns=['_merge'])
```

```python
# merge df_ords_prods_combined and df_prods on product_id

df_ords_prods_merge = df_ords_prods_combined.merge(df_prods, on = 'product_id', indicator = True)
```
